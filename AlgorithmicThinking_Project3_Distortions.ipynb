```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df_111 = pd.read_csv('distortions_111.csv')
df_290 = pd.read_csv('distortions_290.csv')
df_896 = pd.read_csv('distortions_896.csv')
```

```
plt.plot(df_111['n_out_clusters'], df_111['h_clustering'], label='Hierarchical clustering')
plt.plot(df_111['n_out_clusters'], df_111['k_m_clustering'], label='K-Means clustering')
plt.xlabel('No. of output clusters')
plt.ylabel('Distortion (in 10^11 units)')
plt.title('Distortions on the 111 Counties Dataset', fontdict={'fontsize': 15}, pad=22)
plt.legend()
plt.savefig('distortions_111.jpg', quality=100, dpi=200, pad_inches=1)
plt.show()
```

```
plt.plot(df_290['n_out_clusters'], df_290['h_clustering'], label='Hierarchical clustering')
plt.plot(df_290['n_out_clusters'], df_290['k_m_clustering'], label='K-Means clustering')
plt.xlabel('No. of output clusters')
plt.ylabel('Distortion (in 10^12 units)')
plt.title('Distortions on the 290 Counties Dataset', fontdict={'fontsize': 15}, pad=22)
plt.legend()
plt.savefig('distortions_290.jpg', quality=100, dpi=200, pad_inches=1)
plt.show()
```
